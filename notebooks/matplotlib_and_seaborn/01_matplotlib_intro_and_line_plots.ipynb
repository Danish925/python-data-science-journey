# 📉 Part 1: Matplotlib Introduction and Line Plot Fundamentals

**Goal:** Understand the fundamental architecture of Matplotlib (Figure vs. Axes) and master the basics of creating and customizing line plots using Pandas Series and DataFrame columns.

---
### Key Learning Objectives
1.  Understand the difference between the **Pyplot** (`plt.`) and **Object-Oriented (OO)** (`ax.`) interfaces.
2.  Learn how to plot a Pandas **Series** (index as X-axis) and **DataFrame columns** (explicit X/Y).
3.  Customize line appearance using colors, markers, and line styles.
4.  Use **Subplots** to display multiple plots on one Figure.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Set pandas display options
pd.set_option('display.max_columns', None)
pd.set_option('display.precision', 2)

print("LAB 1.1: ENVIRONMENT SETUP.")
print("✅ Matplotlib and pandas imported successfully!")
print(f"Matplotlib version: {plt.matplotlib.__version__}")
print(f"Pandas version: {pd.__version__}\n")
```

## 2. Your First Plot (Basic Pyplot Interface)

The simplest way to plot data is using the **Pyplot** interface (`plt.method()`). We pass Pandas Series directly as the X and Y inputs.

```python
print("LAB 1.2: FIRST PLOT WITH PANDAS DATA.")

# Create simple data using pandas
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 6, 8, 10]
})

# Create and display the plot
plt.plot(data['x'], data['y'])
plt.title("First Plot")
plt.show()

print("\n🎉 You've created your first matplotlib plot with pandas data!")
```

## 3. Figure vs. Axes Concepts (Object-Oriented API)

The **Object-Oriented (OO)** approach gives you more control and is mandatory for complex plots.

* **Figure:** The entire canvas (the outer container).
* **Axes:** The individual plot area (the box with X/Y axes).

Use **`fig, ax = plt.subplots()`** to initialize the OO interface.

```python
print(f"\nLAB 1.3: FIGURE VS AXES CONCEPTS WITH PANDAS.")

# Create sample pandas data
df = pd.DataFrame({
    'day': [1, 4, 2, 3],
    'value': [1, 4, 2, 3]
})

# Method 2: Object-oriented interface (Preferred Method)
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(df['day'], df['value'])
ax.set_title("Plot using OO interface (ax.method)")
ax.set_xlabel("Day")
ax.set_ylabel("Value")
plt.show()

print("📚 Key Concepts:")
print("• Figure = The entire window/canvas")
print("• Axes = The actual plot area with x and y axes")
```

## 4. Line Plot Fundamentals

Line plots track change over an ordered sequence (like time or index).

* **Plotting a Series:** If you pass one Pandas Series, Matplotlib automatically uses the Series **index** for the X-axis.
* **Plotting a DataFrame:** You must explicitly pass two columns for the X and Y axes.

```python
print("LAB 1.4: LINE PLOT FUNDAMENTALS WITH PANDAS.")

# Create a pandas Series 
steps_data = pd.Series([10, 25, 30, 35, 20, 15, 25], 
                        name='Daily Steps (thousands)')

plt.figure(figsize=(8, 5))
plt.plot(steps_data)
plt.title("Single Series Plot (X-values auto-generated)")
plt.xlabel("Day Index")
plt.ylabel(steps_data.name)
plt.show()

print("📝 Notice: x-axis uses pandas Series index automatically!")

# Create a DataFrame
weekly_data = pd.DataFrame({
    'day': [1, 2, 3, 4, 5, 6, 7],
    'steps': [8000, 12000, 6000, 15000, 9000, 11000, 7000],
    'day_name': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
})

plt.figure(figsize=(10, 6))
plt.plot(weekly_data['day'], weekly_data['steps'])
plt.title("Daily Step Count Over a Week")
plt.xlabel("Day of Week")
plt.ylabel("Steps")
plt.grid(True)
plt.show()
```

## 5. Basic Line Customization and Subplots

Customizing a line's color, style, and markers enhances readability. We use **`plt.subplot()`** to place multiple Axes (plots) inside a single Figure.

```python
print(f"\nLAB 1.5: BASIC LINE CUSTOMIZATION WITH PANDAS DATA.")

# Use the same DataFrame from before
weekly_data = pd.DataFrame({
    'day': [1, 2, 3, 4, 5, 6, 7],
    'steps': [8000, 12000, 6000, 15000, 9000, 11000, 7000],
    'day_name': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
})

plt.figure(figsize=(12, 8)) # One Figure container

# Plot 1: Red solid line
plt.subplot(2, 2, 1) # (rows, columns, plot_index)
plt.plot(weekly_data['day'], weekly_data['steps'], color='red')
plt.title("Red Solid Line")

# Plot 2: Blue dashed line
plt.subplot(2, 2, 2)
plt.plot(weekly_data['day'], weekly_data['steps'], color='blue', linestyle='--')
plt.title("Blue Dashed Line")

# Plot 3: Green line with markers
plt.subplot(2, 2, 3)
plt.plot(weekly_data['day'], weekly_data['steps'], color='green', marker='o')
plt.title("Green Line with Markers")

# Plot 4: Purple line, square markers, dash-dot style
plt.subplot(2, 2, 4)
plt.plot(weekly_data['day'], weekly_data['steps'], color='purple', marker='s', linestyle='-.')
plt.title("Purple Line, Square Markers, Dash-dot")

plt.tight_layout() # Adjusts spacing between subplots
plt.show()

print("🎨 Customization syntax:")
print("• Color: `color='red'` or `color='#FF5733'`")
print("• Line Style: `linestyle='--'` (dashed) or `linestyle=':'` (dotted)")
print("• Markers: `marker='o'` (circle) or `marker='s'` (square)")
```

## 6. Practice Project: Titanic Age Profile

We combine all concepts learned (Pandas data handling, line plots, customization, and annotations) to create a professional plot.

```python
print("🎯 PRACTICE PROJECT: Titanic Age Visualization")

# Sample data mimicking the first 10 passengers
titanic_sample = pd.DataFrame({
    'Age': [22, 38, 26, 35, 35, 54, 2, 27, 14, 4],
    'PassengerId': range(1, 11)
})

ages_subset = titanic_sample['Age']
passenger_numbers = titanic_sample['PassengerId']

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(passenger_numbers, ages_subset, color='steelblue', marker='o', 
         linewidth=2, markersize=8)

plt.title('First 10 Titanic Passengers - Age Profile', fontsize=16, fontweight='bold')
plt.xlabel('Passenger Number', fontsize=12)
plt.ylabel('Age (Years)', fontsize=12)
plt.grid(True, alpha=0.3)

# Find and annotate the youngest passenger (using Pandas methods)
youngest_idx = ages_subset.idxmin()
youngest_age = ages_subset.min()
youngest_passenger = passenger_numbers[youngest_idx]

plt.annotate(f'Youngest: {youngest_age} years', 
             xy=(youngest_passenger, youngest_age),
             xytext=(youngest_passenger + 1.5, youngest_age + 5), # Offset the text
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=10, ha='left') # Use ha='left' for clean alignment

plt.tight_layout()
plt.savefig('matplotlib_part1_age_profile.png', dpi=300, bbox_inches='tight')
print("💾 Plot saved as 'matplotlib_part1_age_profile.png'")
plt.show()

# Save the plot

print("🎉 Part 1 complete! Ready for Histograms and Univariate EDA in the next session!")
```
